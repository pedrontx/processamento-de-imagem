```python
#Importando a biblioteca Numpy
import   numpy   as   np
#Importando a biblioteca matplolib
import   matplotlib.pyplot   as   plt
#importando uma biblioteca pil (Python Imaging Library)
from PIL   import   Image 
```

```python
#abrindo a imagem
imagem = Image.open("imagem.png")
#convertendo a imagem para RGB
imagem = imagem.convert("RGB")
#transformando a imagem em um matriz
imagem = np.asarray(imagem)

```

```python
imagem
```

# Espelhamento de imagem 

```python
# inverter a ordem dos elementos 
matriz_espelhada = np.flip(imagem, axis=0)

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.imshow(imagem)
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(matriz_espelhada)
plt.title('Imagem espelhada')
plt.axis('off')

plt.show()
```

# Deslocamento de Imagem

```python
# Definindo o deslocamento temporal
deslocamento = 500  # Altere esse valor conforme necessário
# Realizando o deslocamento temporal
imagem_deslocada = np.roll(imagem, deslocamento, axis=1)
# Exibindo a imagem original e a imagem deslocada
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1) 
plt.imshow(imagem)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_deslocada)
plt.title('Imagem Deslocada')
plt.axis('off')
plt.show()
```

# Escalonamento 

```python
import cv2

nova_largura = 500
nova_altura = 800
imagem_redimensionada = cv2.resize(imagem, (nova_largura, nova_altura))

# Exibindo as imagens lado a lado
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1) 
plt.imshow(imagem)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_redimensionada)
plt.title('Imagem Escalonada')
plt.axis('off')
plt.show()
```
