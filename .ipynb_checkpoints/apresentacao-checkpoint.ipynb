```python
from PIL import Image
import numpy as np

# Abrir a imagem e converter para RGB
imagem = Image.open("imagem.png")
imagem = imagem.convert("RGB")

# Converter a imagem em uma matriz NumPy
imagem_array = np.array(imagem)

# Dicionário de mapeamento de cores para códigos numéricos
cor_para_codigo = {
    (0, 0, 0): 0,       # Preto
    (255, 255, 255): 1, # Branco
    (255, 255, 0): 2,   # Amarelo
    (255, 0, 0): 3,     # Vermelho
    (0, 0, 255): 4      # Azul
    # Adicione outras cores conforme necessário
}

# Função para obter o código numérico da cor
def obter_codigo_cor(cor):
    return cor_para_codigo.get(tuple(cor), -1)  # Retorna -1 se a cor não estiver mapeada

# Aplicar a função de mapeamento a cada pixel na matriz da imagem
codigo_imagem = np.apply_along_axis(obter_codigo_cor, 2, imagem_array)

# Identificar todas as cores únicas presentes na imagem com seus códigos numéricos
cores_presentes = {}

# Percorrer cada pixel na matriz de códigos
altura, largura = codigo_imagem.shape[:2]
for i in range(altura):
    for j in range(largura):
        cor = tuple(imagem_array[i, j])
        codigo = codigo_imagem[i, j]
        if codigo != -1:  # Ignorar códigos não mapeados
            if cor not in cores_presentes:
                cores_presentes[cor] = codigo

# Imprimir as cores presentes na imagem com seus códigos numéricos
print("Cores presentes na imagem com seus códigos numéricos:")
for cor, codigo in cores_presentes.items():
    print(f"Cor: {cor} - Código Numérico: {codigo}")
```

# Espelhamento de imagem 

```python
# inverter a ordem dos elementos 
matriz_espelhada = np.flip(imagem, axis=1)

plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
plt.imshow(imagem)
plt.title('Imagem Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(matriz_espelhada)
plt.title('Imagem espelhada')
plt.axis('off')

plt.show()
```

# Deslocamento de Imagem

```python
# Definindo o deslocamento temporal
deslocamento = 300  # Altere esse valor conforme necessário
# Realizando o deslocamento temporal
imagem_deslocada = np.roll(imagem, deslocamento, axis=1)
# Exibindo a imagem original e a imagem deslocada
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1) 
plt.imshow(imagem)
plt.title('Imagem Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(imagem_deslocada)
plt.title('Imagem Deslocada')
plt.axis('off')
plt.show()
```
